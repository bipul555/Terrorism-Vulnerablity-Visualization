PROGRAM FOR LABEL ENCODING CATEGORICAL VALUES AND DOWNLOADING THEM

```python
import pandas as pd
import io
data1 = pd.read_csv('new_terror_test_data4.csv' , encoding = 'cp437')
print(data1)
```

```python
data1.dtypes
```

```python
data1['provstate'] = data1['provstate'].astype('category')
data1['city'] = data1['city'].astype('category')
data1['gname'] = data1['gname'].astype('category')
data1.dtypes
```

Changing categorical values from char type to 'category' type

```python
data1['provstate'] = data1['provstate'].cat.codes
data1['city'] = data1['city'].cat.codes
data1['gname'] = data1['gname'].cat.codes
print(data1)
```

Downloading the label encoded dataset

```python
from google.colab import files
f8 = open('terror_trial6.csv', 'w')
data1.to_csv(f8, sep = ',', encoding = 'utf-8')
files.download('terror_trial6.csv')
```
